# All

```python
graph = generate_network_PA(level_f='./', label='all')
diffuse_behavior_PA(graph, years=1, thres_PA=p_all[0], I_PA=p_all[1])

PA_results = {}
for node in graph.nodes():
    PA_results[node] = graph.nodes[node]['PA_hist']

PA_df = pd.DataFrame(PA_results).T

print(p_all)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=((14,10)))

PA_df.T.plot(legend=False, ax=axes[0])
axes[0].set_xlabel('Days')
axes[0].set_ylabel('Individual PALs')

PA_df.mean().plot(legend=False, ax=axes[1])
axes[1].set_xlabel('Days')
axes[1].set_ylabel('Mean PAL')
```
