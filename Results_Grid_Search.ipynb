```python
from codes.sim_anneal import grid_search
from codes.networkPA import generate_network_PA
from codes.simulatePA import diffuse_behavior_PA

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d, Axes3D
import numpy as np

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

# All questions Grid Search

```python
list_solutions_all = grid_search(bins=10, label='all', level_f='./')
```

```python
all_df = pd.DataFrame(list_solutions_all).pivot(index=0,columns=1,values=2)
all_df.columns = np.around(np.arange(0.0, 1.0, 0.1), decimals=1)

all_df.index = np.around(np.arange(0.0, 1.0, 0.1), decimals=1)
all_df
```

```python
plt.figure(figsize=((14,12)))
plt.title('Grid search for all questions')
sns.heatmap(all_df, annot=True, annot_kws={"size": 7}, vmin=0, vmax=10)
plt.ylabel('Threshold')
plt.xlabel('I PA')
```
