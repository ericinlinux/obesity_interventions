```python
def generate_environment(level_f):
    '''
    The environment variable is going to be generated randomly by now, but should be replaced later
    '''
    e = pd.read_csv(level_f, sep=';', header=0)
    e=e[['Child_Bosse','School', 'Class','Wave','GEN_FAS_computer_R','GEN_FAS_car_R','GEN_FAS_vacation_R','GEN_FAS_ownroom_R']]
    classes=[67, 71, 72, 74, 77, 78, 79, 81, 83, 86, 100, 101, 103, 121, 122, 125, 126, 127, 129, 130, 131, 133, 135, 136, 138, 139]
    e = e[e['Class'].isin(classes)]
    df_help_missing_e=e.groupby('Child_Bosse').sum()
    df_help_missing_e=df_help_missing_e.drop(['School', 'Class','Wave'], axis=1)
    df_only_missing=df_help_missing_e[df_help_missing_e.isnull().sum(1)==4]
    missing_users_index=df_only_missing.index.values
    users_with_data_e = e[~e['Child_Bosse'].isin(missing_users_index.tolist())]
    index_kids=users_with_data_e.Child_Bosse.unique().tolist()
    user_list_same_waves_val = [] 
    user_list_diff_waves_val = [] 
    for user in index_kids:
        user_df=users_with_data_e[users_with_data_e.Child_Bosse == user]
        if(user_df.GEN_FAS_computer_R.nunique()==1 and user_df.GEN_FAS_car_R.nunique()==1 and user_df.GEN_FAS_vacation_R.nunique()==1 and user_df.GEN_FAS_ownroom_R.nunique()==1):
            user_list_same_waves_val.append(user)
        else:
            user_list_diff_waves_val.append(user)

    final_e=users_with_data_e[users_with_data_e.Wave==1]
    col_list= list(['GEN_FAS_computer_R','GEN_FAS_car_R', 'GEN_FAS_vacation_R','GEN_FAS_ownroom_R'])
    
    #ncol=final_e[col_list].sum(axis=1).to_frame()
    final_e['FAS_Score_R'] = final_e['GEN_FAS_computer_R'] + final_e['GEN_FAS_vacation_R'] + \
            final_e['GEN_FAS_car_R']*1.5 + final_e['GEN_FAS_ownroom_R']*3
    
    #ncol.columns=['FAS_Score_R']
    #final_e = pd.concat([final_e, ncol], axis=1)
    final_e=final_e[['Child_Bosse','FAS_Score_R']]
    final_e.set_index('Child_Bosse')
    #final dataframe containing childID and FAS_Score_R

    return final_e

```

```python
def pa_regression(level_f,pa_type):
    fitbit = pd.read_csv(level_f, sep=';', header=0)
    if pa_type=='Steps_ML_imp1':
        fitbit=(fitbit.groupby(['Child_Bosse','Wave']).mean()[pa_type]* 0.000153).to_frame()
    elif pa_type=='Minutes_MVPA_ML_imp1':
        fitbit=fitbit.groupby(['Child_Bosse','Wave']).mean()[pa_type].to_frame()
    # remove Wave from index    
    fitbit=fitbit.reset_index(level=[1])
    # map the appropriate Wave values, they represent months
    fitbit['Wave']=fitbit['Wave'].map({1:1, 2:3, 3:6, 4:12})
    # get all the unique users
    users=np.unique(fitbit.index.values)
    # remove ChildID from index 
    fitbit.reset_index(level=0, inplace=True)
    
    for userID in users:
        #'help' df for a particular user
        u=fitbit[fitbit.Child_Bosse==userID]
        # get X axis values - the months
        user_X=u['Wave']
        # get Y axis values - the physical activity measurements
        user_Y=u[pa_type]
        user_X = np.asarray(user_X)
        user_Y = np.asarray(user_Y)
        # implement the linear regression
        fit = np.polyfit(user_X, user_Y, 1)
        # fit_fn holds the datapoints from the linear regression
        fit_fn = np.poly1d(fit)
        # replace the old pa values with the new values obtained from the linear regression
        fitbit.loc[(fitbit.Child_Bosse==userID) & (fitbit.Wave==1),pa_type]=fit_fn(1)
        fitbit.loc[(fitbit.Child_Bosse==userID) & (fitbit.Wave==3),pa_type]=fit_fn(3)
        fitbit.loc[(fitbit.Child_Bosse==userID) & (fitbit.Wave==6),pa_type]=fit_fn(6)
        fitbit.loc[(fitbit.Child_Bosse==userID) & (fitbit.Wave==12),pa_type]=fit_fn(12)
    
    return fitbit
```
