## Random

```python
g_all_10_list = []
g_gen_10_list = []
g_friend_10_list = []

g_all_15_list = []
g_gen_15_list = []
g_friend_15_list = []

g_all_20_list = []
g_gen_20_list = []
g_friend_20_list = []

for i in range(100):
    print('\n+++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    print(i)
    print('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n')
    # 10%
    g_all_10 = apply_intervention_random_nodes(graph_all.copy(), perc=0.1)
    diffuse_behavior_PA(graph=g_all_10, thres_PA=parameters_all[0], I_PA=parameters_all[1], years=1)
    g_all_10_list.append(get_graphs_PA_df(g_all_10))

    g_gen_10 = apply_intervention_random_nodes(graph_gen.copy(), perc=0.1)
    diffuse_behavior_PA(graph=g_gen_10, thres_PA=parameters_gen[0], I_PA=parameters_gen[1], years=1)
    g_gen_10_list.append(get_graphs_PA_df(g_gen_10))

    g_friend_10 = apply_intervention_random_nodes(graph_friend.copy(), perc=0.1)
    diffuse_behavior_PA(graph=g_friend_10, thres_PA=parameters_friend[0], I_PA=parameters_friend[1], years=1)
    g_friend_10_list.append(get_graphs_PA_df(g_friend_10))


    # 15%
    g_all_15 = apply_intervention_random_nodes(graph_all.copy(), perc=0.15)
    diffuse_behavior_PA(graph=g_all_15, thres_PA=parameters_all[0], I_PA=parameters_all[1], years=1)
    g_all_15_list.append(get_graphs_PA_df(g_all_15))

    g_gen_15 = apply_intervention_random_nodes(graph_gen.copy(), perc=0.15)
    diffuse_behavior_PA(graph=g_gen_15, thres_PA=parameters_gen[0], I_PA=parameters_gen[1], years=1)
    g_gen_15_list.append(get_graphs_PA_df(g_gen_15))

    g_friend_15 = apply_intervention_random_nodes(graph_friend.copy(), perc=0.15)
    diffuse_behavior_PA(graph=g_friend_15, thres_PA=parameters_friend[0], I_PA=parameters_friend[1], years=1)
    g_friend_15_list.append(get_graphs_PA_df(g_friend_15))

    # 20%
    g_all_20 = apply_intervention_random_nodes(graph_all.copy(), perc=0.2)
    diffuse_behavior_PA(graph=g_all_20, thres_PA=parameters_all[0], I_PA=parameters_all[1], years=1)
    g_all_20_list.append(get_graphs_PA_df(g_all_20))

    g_gen_20 = apply_intervention_random_nodes(graph_gen.copy(), perc=0.2)
    diffuse_behavior_PA(graph=g_gen_20, thres_PA=parameters_gen[0], I_PA=parameters_gen[1], years=1)
    g_gen_20_list.append(get_graphs_PA_df(g_gen_20))

    g_friend_20 = apply_intervention_random_nodes(graph_friend.copy(), perc=0.2)
    diffuse_behavior_PA(graph=g_friend_20, thres_PA=parameters_friend[0], I_PA=parameters_friend[1], years=1)
    g_friend_20_list.append(get_graphs_PA_df(g_friend_20))
```
